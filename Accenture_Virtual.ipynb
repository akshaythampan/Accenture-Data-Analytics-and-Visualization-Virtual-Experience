```python
import pandas as pd
import matplotlib.pyplot as plt
import re
import seaborn as sns
```

```python
# Replace 'file_name.csv' with the actual name of your CSV file
file_path = 'Cleaned Dataset - Sheet1.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)
```

```python
df.dtypes
```

```python
df.head(10)
```

```python
# Define a function to remove double quotes from a Category Column
def remove_double_quotes(value):
    return re.sub(r'"', '', value)

# Apply the function to the 'cate' column
df['Category'] = df['Category'].apply(remove_double_quotes)
```

```python
# Get the value counts for each unique value in the column
value_counts = df['Content ID'].value_counts()

# Find the most frequent value
most_frequent_value = value_counts.idxmax()

# Find the count of the most frequent value
most_frequent_count = value_counts.max()

print("Most frequent value:", most_frequent_value)
print("Count of the most frequent value:", most_frequent_count)
```

```python
# Get the value counts for each unique value in the column
value_counts = df['Category'].value_counts()

# Get the number of unique values in the column
num_unique_values = len(value_counts)

print("Number of unique values:", num_unique_values)
print("\nValue Counts:")
print(value_counts)
```

```python
# Get the value counts for each unique value in the Category column
value_counts = df['Category'].value_counts()

# Get the number of unique values in the column
num_unique_values = len(value_counts)

print("Number of unique values:", num_unique_values)
print("\nValue Counts:")
print(value_counts)
```

```python
# Get the value counts for each unique value in the Type column
value_counts = df['Type'].value_counts()

# Get the number of unique values in the column
num_unique_values = len(value_counts)

print("Number of unique values:", num_unique_values)
print("\nValue Counts:")
print(value_counts)
```

```python
# Get the value counts for each unique value in the Sentiments column
value_counts = df['Sentiments'].value_counts()

# Get the number of unique values in the column
num_unique_values = len(value_counts)

print("Number of unique values:", num_unique_values)
print("\nValue Counts:")
print(value_counts)
```

```python
# Get the value counts for each unique category in the 'Category' column
category_counts = df['Category'].value_counts().head(5)

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)

# Add a title
plt.title("POPULARITY PERCENTAGE SHARE FROM TOP 5 CATEGORIES")

# Display the pie chart
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()

```

```python
# Get the value counts for each unique value in the 'Category' column
value_counts = df['Category'].value_counts()

# Get the top 5 values with the highest counts
top_5_values = value_counts.head(5)

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_5_values, labels=top_5_values.index, autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Categories')
plt.legend(title='Categories', loc='best')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Display the pie chart
plt.show()
```

```python
# Get the value counts for each unique category in the 'Category' column
category_counts = df['Category'].value_counts()

# Sort the categories in descending order based on their counts
sorted_categories = category_counts.sort_values(ascending=False)

# Plot the bar chart
plt.figure(figsize=(10, 6))
colors = plt.cm.Set3(range(len(sorted_categories)))  # Choose a colormap for the bars

# Plot the bars with colors and labels
plt.bar(sorted_categories.index, sorted_categories, color=colors)

# Customize the plot
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Category Distribution with Sentiments')
plt.xticks(rotation=45, ha='right')  # Rotate the x-axis labels for better visibility
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Create a cross-tabulation (contingency table) between 'Category' and 'Type'
cross_tab = pd.crosstab(df['Category'], df['Type'])

# Create the heatmap using seaborn
plt.figure(figsize=(10, 6))
sns.heatmap(cross_tab, cmap='YlGnBu', annot=True, fmt='d')
plt.xlabel('Type')
plt.ylabel('Category')
plt.title('Heatmap: Category vs. Type')
plt.tight_layout()
plt.show()
```

```python
# Get the value counts for each sentiment category
sentiment_counts = df['Sentiments'].value_counts()

# Create a bar chart
plt.figure(figsize=(6, 4))
colors = ['green', 'orange', 'red']  # Define colors for positive, neutral, and negative sentiments
plt.bar(sentiment_counts.index, sentiment_counts, color=colors)

# Customize the plot
plt.xlabel('Sentiments')
plt.ylabel('Count')
plt.title('Sentiment Distribution')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()

```
